```python
import os
import pandas as pd
import numpy as np
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

## LOAD DATASET | AUTO GENERATED DATSET

```python
df = pd.read_csv("salary_data_large.csv")
```

### Data Summary and missing value

```python
df = df.dropna()
df = df[df['salary'] > 5000]         # Remove wrong data
df = df[df['salary'] < 500000]
```

### Prepare Features

```python
X = df[['experience', 'job_role', 'location']]
y = df['salary']
```

### ENCODING

```python
encoder = OneHotEncoder(handle_unknown='ignore')
X_encoded = encoder.fit_transform(X).toarray()
```

### SAVING ENCODER

```python
os.makedirs("model", exist_ok=True)
pickle.dump(encoder, open("model/encoder.pkl", "wb"))
```

### Train / Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)
```

### Train Model

```python
model = RandomForestRegressor(
    n_estimators=400,
    max_depth=25,
    random_state=42
)

model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
accuracy
```

### Save the model

```python
pickle.dump(model, open("model/salary_model.pkl", "wb"))

print("Model saved successfully!")
print("Accuracy:", accuracy)
```
